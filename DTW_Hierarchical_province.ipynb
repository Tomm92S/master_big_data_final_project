### DTW warping path function

```python
DTW = df 

def DTW_dist(DTW, dist = 0, i=DTW.shape[0]-1, j=DTW.shape[1]-1):
    
    value = DTW[i,j]
    a = DTW[i,j-1]
    b = DTW[i-1,j]     
    c = DTW[i-1,j-1]
        
    DTW_min = min(a,b,c)
    dist += DTW_min
    ij_min = np.where(DTW == DTW_min)
    last_i = i
    
    i = ij_min[0][(ij_min[0] == i) | (ij_min[0] == i-1)][0]
    
    if i == last_i:
        j = j-1
    else:
        j = ij_min[1][(ij_min[1] == j) | (ij_min[1] == j-1)][0]
    
    if i == 0 and j == 0:
 
        return dist
    
    return DTW_dist(DTW, dist = dist, i=i, j=j)
```

```python
%%time
total_iter = 0
count1 = 0
lista = []
w = 7
total = (df.shape[1]*(df.shape[1]-1))/2
upper_side_distance_matrix = []

for series1 in df.T:
    count1+=1
    count2 = 0
    for series2 in df.T:
        count2+=1
        if count1 == count2:
            continue
            
        if (count1,count2) in lista:
            continue      
        lista.append((count1,count2))
        lista.append((count2,count1))
        total_iter+=1
        
        print(round((total_iter/total*100),2),'%')
        
        #da qui inizia il codice
        n = df.shape[0] + 1
        m = n
        DTW = np.zeros([n,m])
        DTW[0,1:] = np.inf
        DTW[1:,0] = np.inf

        a = series1.reshape(1,df.shape[0])
        b = series2.reshape(1,df.shape[0])
        
        #calcolo DISTANCE MATRIX (con window = 6)
        #w = 6
        for i in range(1,n):
            for j in range(max([1,i-w]),min([m, i+w])):
                dist = abs(a[:,i-1] - b[:,j-1]) 
                DTW[i,j] = dist + min([DTW[i-1,j], DTW[i,j-1], DTW[i-1,j-1]])
        
        #calcolo WARPING PATH
        upper_side_distance_matrix.append(DTW_dist(DTW, dist = 0, i=DTW.shape[0]-1, j=DTW.shape[1]-1))   

#Salviamo la DISTANCE MATRIX in un file.txt
data_dist = np.array(upper_side_distance_matrix)
np.savetxt("data_dist_province.txt", data_dist)
```

### Plot with sigma

```python
def plot_std(df, labels, x_axis = [0], start_one = False, plot_std = True, **kwargs, ):
    
    if start_one == True:
        dfn = df.T.copy()
        for series in range(dfn.shape[0]):
            #print(series)
            for i in range(dfn.shape[1]):
                #print(i)
                dfn[series][i] = df.T[series][i] - df.T[series][0]+1       
        dfl = np.append(dfn.T,labels.reshape(1,dfn.shape[0]), axis = 0).T
        
    else:       
        dfl = np.append(df,labels.reshape(1,df.shape[1]), axis = 0).T
        
        
    fig,ax = plt.subplots(1,1,figsize=(12,8))
    #colors = ['g','orange','b','#cc0099','r']
    colors = ['g','r']
    set_link_color_palette(['#66c2a5','#fc8d62','#8da0cb'])
    index = False
    if x_axis != [0]:
        index = True
        
    #len(np.unique(labels))
    for label in range(len(np.unique(labels))):
        label+=1    

        meancentr = np.mean(dfl[dfl[:,-1] == label][:,:-2],axis = 0)  

        mu = dfl[dfl[:,-1] ==label][:,:-2].mean(0)
        std = dfl[dfl[:,-1] ==label][:,:-2].std(0)

        centroid = ax.plot(meancentr, c = colors[label-1], **kwargs) #
        if index == False:
            x_axis = [c for c in range(meancentr.shape[0])]
        
        if plot_std == True:
            ax.fill_between(x_axis, mu - std, mu + std, edgecolor='none',
                            facecolor=centroid[0].get_color(), alpha=0.2)
    ax.autoscale(axis='x',tight = True)
    ax.set_title('DTW Hierarchical clustering centroids (with STD)', fontsize = 20)
    #ax.set_xlabel('month')
    ax.set_ylabel('Standardized AVG prices',fontsize=13)
    plt.xticks(ticks=[7,19,31,43,55,67,79,91],labels=['2013','2014','2015','2016','2017','2018','2019','2020'],fontsize=13)
    #ax.legend(loc = 0)
```

```python
%%time
dist_matrix = np.zeros(df.shape[1]*df.shape[1]).reshape(df.shape[1],df.shape[1])

#UPPER
c = 0
left = 0
for row in range(0,dist_matrix.shape[0]-1):
    left+=1
    for col in range(left,dist_matrix.shape[1]):
        dist_matrix[row,col] = data_dist[c]
        c+=1
        
#LOWER
c = 0
right = 0
for row in range(1,dist_matrix.shape[0]):
    right+=1
    for col in range(0,right):
        dist_matrix[row,col] = data_dist[c]
        c+=1
```

### SSE

```python
def SSE(df, labels):
    
    
    dfl = np.append(df,labels.reshape(1,df.shape[1]), axis = 0).T 
    
    # for every cluster...
    SSE = 0
    for label in range(len(np.unique(labels))):
        label+=1    

        meancentr = np.mean(dfl[dfl[:,-1] == label][:,:-2],axis = 0)
        n = len(meancentr) +1
        
        
        cluster_sum = 0
        # for every series in the cluster...
        for series_index in range(len(dfl[dfl[:,-1] == label])):
            
            #...compute the DTW**2 between the series and the centroid            
            series = dfl[dfl[:,-1] == label][series_index,:-2]
            m = len(series) +1
            DTW = np.zeros([n,m])
            DTW[0,1:] = np.inf
            DTW[1:,0] = np.inf
            
            a = meancentr.reshape(1,len(meancentr))
            b = series.reshape(1,len(series))
            
            #calcolo DISTANCE MATRIX (con window = 6)
            #w = 6
            for i in range(1,n):
                for j in range(max([1,i-w]),min([m, i+w])):
                    dist = abs(a[:,i-1] - b[:,j-1]) 
                    DTW[i,j] = dist + min([DTW[i-1,j], DTW[i,j-1], DTW[i-1,j-1]])
                    
            #calcolo WARPING PATH
            cluster_sum += (DTW_dist(DTW, dist = 0, i=DTW.shape[0]-1, j=DTW.shape[1]-1))**2
        SSE += cluster_sum
    return SSE
            
            #take the sum
```

### SSE plot

```python
def plot_elbow_graph(data_dist, method = 'complete', thresh_range=(10, 140, 10)):
    sse = {}
    for threshold in range(thresh_range[0],thresh_range[1],thresh_range[2]):
        data_link = linkage(data_dist, method=method)
        labels = fcluster(data_link, t=threshold, criterion='distance')
        nclusters = labels.max()


        sse[nclusters] = SSE(df, labels)

    plt.figure(figsize = (16,10))
    plt.plot(list(sse.keys()), list(sse.values()))
    plt.xlabel("Number of clusters", fontsize = 15)
    plt.ylabel("SSE", fontsize = 15)
    plt.grid(axis='both', linestyle='-')
    plt.xticks(list(sse.keys()))
    plt.yticks(list(sse.values()))
    plt.autoscale(tight = True);
```
