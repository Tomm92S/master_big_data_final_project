# Plot_maps_DTW_Hierarchical_province

```python
# import the library
import pandas as pd
import geopandas as gpd # Let's import geopandas
import matplotlib.pyplot as plt
import numpy as np
```

```python
tess = gpd.read_file("limits_IT_provinces.geojson")
labels = pd.read_csv('province_labels_2.csv')
labels = labels.rename(columns = {'comune':'prov_name'})
tess['prov_name'] = tess['prov_name'].apply(lambda x : x.strip().lower().replace(' ','-'))
tess['prov_name'] = tess['prov_name'].replace({'bolzano/bozen':'bolzano','ascoli-piceno':'ascoli','barletta-andria-trani':'barletta',\
                         'forlì-cesena':'forli',"l'aquila":'aquila','massa-carrara':'massa','monza-e-della-brianza':'monza',\
                         'pesaro-e-urbino':'pesaro','reggio-di-calabria':'reggio-calabria',"reggio-nell'emilia":'reggio',\
                         "valle-d'aosta/vallée-d'aoste":'valle-d-aosta','verbano-cusio-ossola':'verbania',\
                         'vibo-valentia':'vibo'})
print(set(labels['prov_name'])-set(tess['prov_name']))
tess = pd.merge(tess,labels,how = 'left',on = 'prov_name')
tess['label']=tess['label'].replace({1.0:'Upward shape',2.0:'Downward shape'})
```

```python
import matplotlib.colors as clr
cmap = clr.ListedColormap(['r', '#ff1ac6','#111111','#4d4dff',   'pink','orange', '#00b300'])
```
